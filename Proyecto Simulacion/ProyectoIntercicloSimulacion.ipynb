## Introduction

Simulation is the imitation of the functioning of a real system during a time interval. This simulation can be done either manually or computationally. The simulation is based on a model of reality that tells a story and by observing the behavior of this, allows us to obtain knowledge about the real system. The behavior of the simulation is determined by the simulation model or set of assumptions concerning the System Real, these assumptions are expressed through logical relations and Mathematics between entities.

## Stages for simulation
### Definition of the system

It consists of studying the context of the problem, identifying the objectives of the project, specifying the measurement indices of the system's effectiveness, specifying the specific objectives of the modeling and defining the system to be modeled.

### Formulation of the model

Once the results that are expected to be obtained from the study are defined with exactitude, the model with which the desired results will be obtained is defined and constructed. In the formulation of the model, it is necessary to define all the variables that are part of it, its logical relationships and the flow diagrams that describe the model completely.

### Collection of data

It is important that the data that the model will require to produce the desired results be defined with clarity and accuracy.

### Implementation of the model in the computer

With the defined model, the next step is to decide if a language such as fortran, algol, lisp, etc. is used, or a package like Promodel, Vensim, Stella and iThink is used, GPSS, simulates, simscript, Rockwell Arena, [ Flexsim], etc., to process it on the computer and obtain the desired results.
check

The verification process consists in verifying that the simulated model meets the design requirements for which it was developed. It is about evaluating that the model behaves according to its model design
System Validation

Through this stage is to assess the differences between the operation of the simulator and the real system that is trying to simulate. [3] The most common ways to validate a model are:

    The opinion of experts on the results of the simulation.
    The accuracy with which historical data is predicted.
    The accuracy in the prediction of the future.
    The failure check of the simulation model when using data that makes the real system fail.
    The acceptance and trust in the model of the person who will make use of the results that the simulation experiment yields.

###  Experimentation

The experimentation with the model is done after it has been validated. Experimentation consists of generating the desired data and performing a sensitivity analysis of the required indices ...
Interpretation

In this stage of the study, the results of the simulation are interpreted and based on this a decision is made. It is obvious that the results obtained from a simulation study help support decisions of the semi-structured type.

# Tools

SimPy is a process-based discrete-event simulation framework based on standard Python. Its event dispatcher is based on Python’s generators and can also be used for asynchronous networking or to implement multi-agent systems (with both, simulated and real communication).

Processes in SimPy are simple Python generator functions and are used to model active components like customers, vehicles or agents. SimPy also provides various types of shared resources to model limited capacity congestion points (like servers, checkout counters and tunnels). From version 3.1, it will also provide monitoring capabilities to aid in gathering statistics about resources and processes.

```python
import simpy
import random
import matplotlib.pyplot as pp

%matplotlib inline

# maximo de mecanicos
NUM_INGENIEROS_MECANICOS = 10

# Tiempo de atención de sistemas del vehiculodanio
TIEMPO_ATENCION_SISTEMA_FRENOS = random.randint(20,40)
TIEMPO_ATENCION_SISTEMA_TRANSMISION = random.randint(30,60)
TIEMPO_ATENCION_COMPRESION = random.randint(15,30)
TIEMPO_ATENCION_SISTEMA_ELECTRICO= random.randint(30,45)
TIEMPO_ATENCION_ABC= random.randint(25,90)

#Areas

NUM_AREA_A=20

NUM_AREA_B=13

NUM_AREA_C=23

NUM_AREA_D=12

NUM_AREA_ABC = 10

#Numero de cajas para repuestos
NUM_CAJA_REPUESTOS=4


NUM_AREA_REPUESTOS=20

TIEMPO_ATENCION_REPUESTOS= random.randint(10,15)

# Tiempo de simulación
TIEMPO_SIMULACION = 300

#Numero de vehiculodanios que van a llegar
NUM_VEHICULOS = 20

#numero de vehiculodanios por tipo y por danio

NUM_FRENOS = round(0.60 * NUM_VEHICULOS)
print('SISTEMA DE FRENOS GENERADO')
print(NUM_FRENOS)

NUM_TRANSMISION = round(0.35 * NUM_VEHICULOS)
print('SISTEMA DE TRANSMISION GENERADO')
print(NUM_TRANSMISION)

NUM_COMPRESION = round(0.15 * NUM_VEHICULOS)
print('SISTEMA DE COMPRESION GENERADO')
print(NUM_COMPRESION)

NUM_ELECTRICO = round(0.30 * NUM_VEHICULOS)
print('SISTEMA ELECTRICO GENERADO')
print(NUM_ELECTRICO)

NUM_ABC = round(0.10 * NUM_VEHICULOS)
print('REVISION ABC')
print(NUM_ABC)

#generamos grafica de los vehiculodanios creados
fig = pp.figure(u'Gráfica de barras de Vehiculos') # Figure
ax = fig.add_subplot(111) # Axes

nombres = ['S. FRENOS','S. TRANSMISION','S. COMPRESION','S. ELECTRICO','ABC']
datos = [NUM_FRENOS,NUM_TRANSMISION,NUM_COMPRESION,NUM_ELECTRICO,NUM_ABC]
xx = range(len(datos))

ax.bar(xx, datos, width=0.8, align='center')
ax.set_xticks(xx)
ax.set_xticklabels(nombres)

pp.show()

# Intervalo de tiempo en que llegan pasientes
INTERVALO_LLEGADA = 9

tiempofrenos={}
tiempotransmision={}
tiempocompresion={}
tiempoelectrico={}
tiempoabc={}


class AtencionMirasol(object):
    
    def __init__(self, environment, num_ingenieros_mecanicos, tiempo_atencion_sistema_frenos, tiempo_atencion_sistema_transmision, tiempo_atencion_compresion, tiempo_atencion_sistema_electrico, tiempo_atencion_abc,num_area_a, 
                num_area_b, num_area_c, num_area_d,num_area_abc, num_caja_repuestos, num_area_repuestos, tiempo_atencion_repuestos,):
        
        # Guardamos como variable el entorno de ejecucion
        self.env=environment
        # Creamos el recurso que representa el numero de mecanicos
        self.mecanicos = simpy.Resource(environment, num_ingenieros_mecanicos)
    
        self.areafrenos = simpy.Resource(environment, num_area_a)
      
        self.areatransmision = simpy.Resource(environment, num_area_b)
     
        self.areacompresion = simpy.Resource(environment, num_area_c)
    
        self.areaelectrico = simpy.Resource(environment, num_area_d)
        
        self.areaabc = simpy.Resource(environment , num_area_abc)

        self.consultorios = simpy.Resource(environment,num_caja_repuestos)
      
        self.salaEsperaConsultorios = simpy.Resource(environment, num_area_repuestos)
        
        # Variable para el tiempo de atencion 
        self.tiempo_atencion_sistema_frenos = tiempo_atencion_sistema_frenos
        self.tiempo_atencion_sistema_transmision = tiempo_atencion_sistema_transmision
        self.tiempo_atencion_compresion = tiempo_atencion_compresion
        self.tiempo_atencion_sistema_electrico= tiempo_atencion_sistema_electrico
        self.tiempo_atencion_repuestos= tiempo_atencion_repuestos
        self.tiempo_atencion_abc = tiempo_atencion_abc
        

    def atencion_area_revision(self, vehiculodanio):
            print('Para ver que contiene vehiculodanio')
            print(vehiculodanio)    
            tipodanio = vehiculodanio.split('-')
            #condiciones que permitira identificar que tipo de vehiculodanio esta siendo atendido y fue atendido
            if(tipodanio[0] == 'frenos'):
                yield self.env.timeout(TIEMPO_ATENCION_SISTEMA_FRENOS)
                print('frenos ya revisado')

            if(tipodanio[0] == 'Transmision'):
                yield self.env.timeout(TIEMPO_ATENCION_SISTEMA_TRANSMISION)
                print('Transmision ya revisado')

            if(tipodanio[0] == 'Compresion'):
                yield self.env.timeout(TIEMPO_ATENCION_COMPRESION)
                print('Compresion ya revisada')
                
            if(tipodanio[0] == 'Electrico'):
                yield self.env.timeout(TIEMPO_ATENCION_SISTEMA_ELECTRICO)
                print('Electrico ya revisado')
        
        
    
def asignacion_area(env, vehiculodanio, atencionmirasol, tipo):
    # Usamos el reloj de la simulacion (env.now()) para indicar a la
    # nombre pasado como parametro vehiculodanio
    print('Llega Vehiculo: %s con danio %s a la hora %.2f.' % (tipo, vehiculodanio, env.now))

    # Especificamos que vamos a usar un recurso (Resource) que representa
    tipodanio = vehiculodanio.split('-')
    if(tipodanio[0] == 'frenos'):
        with atencionmirasol.areafrenos.request() as areafrenos:
            # Ocupamos la sala de espera frenoss
            yield areafrenos
            #Indicamos que el vehiculodanio se esta trasladando para ser lavado
            print('Se traslada frenos a la sala de revision : %s a la hora %.2f.' % (vehiculodanio,env.now))
            if(tipo == 'singarantia'):
                print('Este vehiculo no cuenta con garantia puede solo comprar repuestos')
                yield env.process(atencionmirasol.atencion_area_revision(vehiculodanio))
            else:
                yield env.process(asignacion_sala_revision(vehiculodanio, atencionmirasol, tipo))
            
            k1=env.now
            if k1 in tiempofrenos:   
                tiempofrenos[k1]=tiempofrenos[k1]+1
            else:
                tiempofrenos[k1]=1
        
    if(tipodanio[0] == 'Transmision'):
        with atencionmirasol.areatransmision.request() as areatransmision:
            # Ocupamos la sala de espera adolesentes
            yield areatransmision
            #Indicamos que el vehiculodanio se esta trasladando para ser lavado
            print('Se traslada Transmision a la sala de revision general: %s a la hora %.2f.' % (vehiculodanio,env.now))
            if(tipo == 'singarantia'):
                print('Este vehiculo no cuenta con garantia puede solo comprar repuestos')
                yield env.process(atencionmirasol.atencion_area_revision(vehiculodanio))
            else:
                yield env.process(asignacion_sala_revision(vehiculodanio, atencionmirasol, tipo))
            
            k2=env.now
            if k2 in tiempotransmision:
                tiempotransmision[k2]=tiempotransmision[k2]+1
            else:
                tiempotransmision[k2]=1

    if(tipodanio[0] == 'Compresion'):
        with atencionmirasol.areacompresion.request() as areacompresion:
            # Ocupamos la sala de espera Compresion
            yield areacompresion
            #Indicamos que el vehiculodanio se esta trasladando para ser lavado
            print('Se traslada Compresion a la sala de revision general: %s a la hora %.2f.' % (vehiculodanio,env.now))
            if(tipo == 'singarantia'):
                print('Este vehiculo no cuenta con garantia puede solo comprar repuestos')
                yield env.process(atencionmirasol.atencion_area_revision(vehiculodanio))
            else:
                yield env.process(asignacion_sala_revision(vehiculodanio, atencionmirasol, tipo))
            
            k3=env.now
            if k3 in tiempocompresion:
                tiempocompresion[k3]=tiempocompresion[k3]+1
            else:
                tiempocompresion[k3]=1
        
    if(tipodanio[0] == 'Electrico'):
        with atencionmirasol.areaelectrico.request() as areaelectrico:
            # Ocupamos la sala de espera Compresion Mayores
            yield areaelectrico
            #Indicamos que el vehiculodanio se esta trasladando para ser lavado
            print('Se traslada Electrico a la sala de revision general: %s a la hora %.2f.' % (vehiculodanio,env.now))
            if(tipo == 'singarantia'):
                print('Este vehiculo no cuenta con garantia puede solo comprar repuestos')
                yield env.process(atencionmirasol.atencion_area_revision(vehiculodanio))
            else:
                yield env.process(asignacion_sala_revision(vehiculodanio, atencionmirasol, tipo))
            
            k4=env.now
            if k4 in tiempoelectrico:
                tiempoelectrico[k4]=tiempoelectrico[k4]+1
            else:
                tiempoelectrico[k4]=1
    if(tipo == 'policia'):
         with atencionmirasol.areaabc.request() as areaabc:
            # Ocupamos la sala de espera Compresion Mayores
            yield areaabc
            
            k5=env.now
            if k5 in tiempoabc:
                tiempoabc[k5]=tiempoabc[k5]+1
            else:
                tiempoabc[k5]=1
                
def asignacion_sala_revision(vehiculodanio, atencionmirasol, tipo):
            
    print('Llega Vehiculo a la sala de revision general: %s a la hora %.2f.' % (vehiculodanio, env.now))
    with atencionmirasol.mecanicos.request() as mecanicos:
        yield mecanicos
        print('Entra vehiculodanio a atencion: %s a la hora %.2f.' % (vehiculodanio, env.now))
        yield env.process(atencionmirasol.atencion_area_revision(vehiculodanio))
        # Una vez que termina la llamada con 'yield', se indica que se ha atendido al vehiculodanio
                
                
        
def ejecutar_simulacion(env, num_ingenieros_mecanicos, tiempo_atencion_sistema_frenos, tiempo_atencion_sistema_transmision, tiempo_atencion_compresion, tiempo_atencion_sistema_electrico,tiempo_atencion_abc, num_area_a, 
                num_area_b, num_area_c, num_area_d,num_area_abc, num_caja_repuestos, num_area_repuestos, tiempo_atencion_repuestos, intervalo):
        
    atencionmirasol=AtencionMirasol(env, num_ingenieros_mecanicos, tiempo_atencion_sistema_frenos, tiempo_atencion_sistema_transmision, tiempo_atencion_compresion, tiempo_atencion_sistema_electrico,tiempo_atencion_abc, num_area_a, 
                num_area_b, num_area_c, num_area_d,num_area_abc, num_caja_repuestos, num_area_repuestos, tiempo_atencion_repuestos)
    # Creamos 5 llegadas de vehiculodanios iniciales
    for i in range(random.randint(1, NUM_FRENOS)):
        env.process(asignacion_area(env, 'frenos-%d'%(i+1),atencionmirasol, random.choice(["garantia", "singarantia", "policia"])))
    for i1 in range(random.randint(1, NUM_TRANSMISION)):
        env.process(asignacion_area(env, 'Transmision-%d'%(i1+1),atencionmirasol, random.choice(["garantia", "singarantia", "policia"])))
    for i2 in range(random.randint(1, NUM_COMPRESION)):
        env.process(asignacion_area(env, 'Compresion-%d'%(i2+1),atencionmirasol, random.choice(["garantia", "singarantia", "policia"])))
    for i3 in range(random.randint(1, NUM_ELECTRICO)):
        env.process(asignacion_area(env, 'Electrico-%d'%(i3+1),atencionmirasol, random.choice(["garantia", "singarantia", "policia"])))
    
    # Ejecutamos la simulacion
    while i<= NUM_FRENOS:
        yield env.timeout(random.randint(intervalo-2, intervalo+2))
        i+=1
        env.process(asignacion_area(env, 'frenos-%d'%(i+1),atencionmirasol,random.choice(["garantia", "singarantia", "policia"])))
    while i1<= NUM_TRANSMISION:
        yield env.timeout(random.randint(intervalo-2, intervalo+2))
        i1+=1
  
        env.process(asignacion_area(env, 'Transmision-%d'%(i1+1),atencionmirasol,random.choice(["garantia", "singarantia", "policia"])))
    while i2<= NUM_COMPRESION:
        yield env.timeout(random.randint(intervalo-2, intervalo+2))
        i2+=1  
        env.process(asignacion_area(env, 'Compresion-%d'%(i2+1),atencionmirasol,random.choice(["garantia", "singarantia", "policia"])))
        
    while i3<= NUM_ELECTRICO:
        yield env.timeout(random.randint(intervalo-2, intervalo+2))
        i3+=1
        env.process(asignacion_area(env, 'Electrico-%d'%(i3+1),atencionmirasol,random.choice(["garantia", "singarantia", "policia"])))

print('Servicio de atención MIRASOL CHEVROLET')
# Inicializamos la semilla aleatoria
random.seed(10)

# Creamos el entorno de simulacion
env=simpy.Environment()
env.process(ejecutar_simulacion(env, NUM_INGENIEROS_MECANICOS, TIEMPO_ATENCION_SISTEMA_FRENOS, TIEMPO_ATENCION_SISTEMA_TRANSMISION, TIEMPO_ATENCION_COMPRESION, TIEMPO_ATENCION_SISTEMA_ELECTRICO,TIEMPO_ATENCION_ABC, NUM_AREA_A,
                               NUM_AREA_B, NUM_AREA_C, NUM_AREA_D,NUM_AREA_ABC, NUM_CAJA_REPUESTOS, NUM_AREA_REPUESTOS, TIEMPO_ATENCION_REPUESTOS, INTERVALO_LLEGADA))

# Ejecutamos el proceso durante el tiempo de simulacion
env.run(until = TIEMPO_SIMULACION)     

# Generamos la grafica 
datosfrenoss=sorted(tiempofrenos.items())
print("Sistema de frenos")
print(datosfrenoss)
x, y =zip(*datosfrenoss)
pp.plot(x,y,linewidth=2,color='red')
pp.scatter(x,y,color='blue')
pp.grid(True)
pp.show()


# Generamos la grafica 
datosadolesentes=sorted(tiempotransmision.items())
print("Sistema de Transmision")
print(datosadolesentes)
x, y =zip(*datosadolesentes)
pp.plot(x,y,linewidth=2,color='red')
pp.scatter(x,y,color='blue')
pp.grid(True)
pp.show()

# Generamos la grafica 
datosCompresion=sorted(tiempocompresion.items())
print("Sistema de Compresion")
print(datosCompresion)
x, y =zip(*datosCompresion)
pp.plot(x,y,linewidth=2,color='red')
pp.scatter(x,y,color='blue')
pp.grid(True)
pp.show()

# Generamos la grafica 
datosElectrico=sorted(tiempoelectrico.items())
print("Sistema Electrico")
print(datosElectrico)
x, y =zip(*datosElectrico)
pp.plot(x,y,linewidth=2,color='red')
pp.scatter(x,y,color='blue')
pp.grid(True)
pp.show()

# Generamos la grafica 
datosAbc=sorted(tiempoabc.items())
print("ABC")
print(datosAbc)
x, y =zip(*datosAbc)
pp.plot(x,y,linewidth=2,color='red')
pp.scatter(x,y,color='blue')
pp.grid(True)
pp.show()
```
